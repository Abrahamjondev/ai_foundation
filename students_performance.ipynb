```python
import pandas as pd
import numpy as np

```

```python
df = pd.read_csv("student_performance.csv")
print("Dataset shape (rows, columns): ", df.shape)
print("============================================================================")
print("column name:", df.columns.tolist())
print("============================================================================")
print("First 5 rows of the dataset:")
print(df.head())
print("============================================================================")
print("Summary description of the dataset")
print(df.describe())
print("============================================================================")
print("Data types and null info")
print(df.info())
```

```python
print("============================================================================")
print("Average score by columns:")
df[["class_participation", "attendance_percentage", "total_score"]].mean()
```

```python
print("============================================================================")
print(" Maximumer sentage  and total score by columns:")
df[["class_participation", "attendance_percentage", "total_score"]].max()
```

```python
print("============================================================================")
print(" Minnimumsentage  and minimum score by columns:")
df[["class_participation", "attendance_percentage", "total_score"]].min()
```

```python
print("============================================================================")
print("Corrolation:")
df[["class_participation", "attendance_percentage", "total_score"]].corr()
```

```python
df["total result"] = df[["class_participation", "attendance_percentage", "total_score"]].sum(axis=1)
# print("tiral result")
df["average score"] = df["total_score"] / 2
df["passed"] = np.where(df["average score"] >= 45, "YES", "NO")
print("New columns preview (result, average, passed)")
print(df[["total result", "average score", "passed"]].head(10))
```

```python
weekly_self_study_hours = df.groupby('weekly_self_study_hours')[["class_participation", "attendance_percentage", "total_score"]].mean()
print("student score by test preperation course:")
print(weekly_self_study_hours)

# student id

studentId = df.groupby("student_id")[["class_participation", "attendance_percentage", "total_score"]].mean()
print("student score by test preperation course:")
print(studentId)

# Who passed vs failed

pass_rate = df["passed"].value_counts(normalize=True)*100
print("Pass rate:\n", pass_rate)
```

```python
#  convert scores to numpy array
scores = df[["class_participation", "attendance_percentage", "total_score"]].values

# Mean, std Deviation
print("Numpy Mean:", np.mean(scores, axis=0))
print("Numpy Std:", np.std(scores, axis=0))

# ? Z-score (normalize)
z_score = (scores - np.mean(scores, axis=0)) / np.std(scores, axis=0)
print("Z-sciore (normalize numbers)")
print(z_score[:5])


```

```python
#  How many student scored above 90 average

top_scores = df[df["average score"] >= 90]
print("Top scored student above 90", len(top_scores))

#  How many studens failed

failed = df[df["passed"] == "No"]
print("Failed from exam:", len(failed))
```
